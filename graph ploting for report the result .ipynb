```python
import matplotlib.pyplot as plt

experimental_setting = ['ADPCM-fixed', 'ADPCM-ORNN', 'ADPCM-OGRU', 'ADPCM-GRU']
stages = ['1st stage', '2nd stage', '3rd stage', 'Last stage']
data = [[31.6, 31, 32.2, 30.9], [36.4, 40.3, 42.5, 39.7], [36.8, 40.5, 42.6, 40.1], [44.6, 45, 44.7, 44.8]]

fig, ax = plt.subplots()
markers = ['o', 's', '*', '^']


for i, setting in enumerate(experimental_setting):
    ax.plot(stages, data[i], label=setting, marker=markers[i])

ax.set_xlabel('Different Test Stage Speech Sample Dataset')
ax.set_ylabel('Performance Using SNR Value in dB')
ax.legend()

ax.set_ylim(bottom=25, top=60)
ax.yaxis.set_ticks(range(25, 50, 5))

plt.savefig("Performance SNR Value.jpeg", dpi=300, bbox_inches='tight')
plt.close()
plt.show()
```
